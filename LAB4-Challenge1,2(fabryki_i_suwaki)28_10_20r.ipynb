```
##Celem tej części ćwiczenia jest obliczenie maksymalnego zysku dla okien aluminiowych i drewnianych. Działania odbyły się na podstawie tabeli przekazanej przez 
##prowadzącego na laboratorium
# Import pulpa
from pulp import *
# Tworzenie zmiennej „prob”, aby zawierała dane problemu.
prob = LpProblem("Zyski_z_okien",LpMaximize)  ## Maximize w celu zmaksymalizowania wyniku.
# Tworzenie zmiennych decyzyjnych o nazwie okna alumioniowe i drewniane
x1 = LpVariable("Okna aluminiowe",0,None,LpInteger)
x2 = LpVariable("Okna drewniane",0,None,LpInteger)


# Dodanie odpowiednich ograniczeń zawartych w treści zadania.
prob += 3000.0*x1 + 5000.0*x2, "Zysk maksymalny"
prob += x1 <= 4.0, "Fabryka 1"
prob += 2.0*x2 <= 12.0, "Fabryka2"
prob += 3.0*x1 + 2.0*x2 <= 18.0, "Fabryka3"

# Dane problemu są zapisywane w pliku .lp
prob.writeLP("Zyski_z_okien")
# Rozwiązanie problemu:
prob.solve()
# Wynik obliczeń zostaje przedstawiony poniżej po kompilacji programu.
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)

# Wyprintowanie zysku maksymalnego:
print("Zysk maksymalny = ", value(prob.objective))

### Optymalnym wynikiem dla zysku maksymalnego jest 36000 zł, przy wyprodukowaniu ilości partii okien przez 3 fabryki.
```

```
### Ta część zadania odpowiada za produkcję dżemu z jabłek
prob = LpProblem("Najlepsza produkcja HURT",LpMaximize)

# Dodanie dwóch zmiennych Dzem i Mus
dzem=LpVariable("Dzem",0) # juz tutaj podajemy ograniczenie ze sa nieujemne
mus=LpVariable("Mus",0)
# Dodanie kolejnych pomocniczych zmiennych.
zuzycie_jablek =LpVariable("Jablka",0)
zuzycie_miodu =LpVariable("Miod",0)

## Stworzenie odpowiedniej funkcji

prob += 16* dzem + 11* mus - 0.3 * zuzycie_jablek - 0.2 * zuzycie_miodu
prob += 8 * dzem+ 2 * mus == zuzycie_jablek, "Zuzycie jablek"
prob += 5 * dzem + 5 * mus == zuzycie_miodu, "Zuzycie miodu"
prob += zuzycie_jablek <= 40, "Maksimum jablek"
prob += zuzycie_miodu <= 25, "Maksimum miodu"


prob.solve()
# Status rozwiązania jest printowany po skompilowaniu programu.
print("Status:", LpStatus[prob.status])
# Każda ze zmiennych jest wypisywana z rozwiązaną optymalną wartością
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Zysk wyniesie {} zł  ".format( value(prob.objective)))

## Jak widać przy takich funkcjach i wartościach przeze mnie wprowadzonych zysk wyniesie 63 zł.
```

```
### Importowanie odpowiednich bibliotek
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd
##Rodzaj stylu
style = {'description_width': 'initial'}

### Poniżej stworzyłem slidery umożliwiające zmiany wartości cen, zysku, zużycia itp. w celu ułatwienia interfejsu, aby nie zmieniać ciągle ręcznie wartości.
cena_jablek_slider = FloatSlider(min=0,max = 1, value= 0.3, description="cena_jablek_slider", style= style )
cena_miodu_slider = FloatSlider(min=0,max = 1,value= 0.2, description="cena_miodu_slider", style= style )
zysk_z_dzemu_slider = FloatSlider(min=0,max = 50, value= 16, description="zysk_z_dzemu_slider", style= style )
zysk_z_musu_slider = FloatSlider(min=0,max = 50, value= 11, description="zysk_z_musu_slider", style= style )
zuzycie_jablek_dzem_slider = FloatSlider(min=0,max = 20,value= 8, description="zuzycie_jablek_dzem_slider", style= style )
zuzycie_jablek_mus_slider = FloatSlider(min=0,max = 20,value= 2, description="zuzycie_jablek_mus_slider", style= style )
zuzycie_miodu_dzem_slider = FloatSlider(min=0,max = 20,value= 5, description="zuzycie_miodu_dzem_slider", style= style )
zuzycie_miodu_mus_slider = FloatSlider(min=0,max = 20,value= 5, description="zuzycie_miodu_mus_slider", style= style )
maksimum_jablek_slider = IntSlider(min=0,max = 200,value= 40, description="maksimum_jablek_slider", style= style )
maksimum_miodu_slider = FloatSlider(min=0,max = 200,value= 25, description="maksimum_miodu_slider", style= style )

### Zdefiniowanie zmiennej hurt_produkcja, zawierającej ceny jabłek, miodu, zyski, zużycia itd.
def hurt_produkcja(cena_jablek = 0.3,
    cena_miodu = 0.2,
    zysk_z_dzemu = 16,
    zysk_z_musu = 11,
    zuzycie_jablek_dzem = 8,
    zuzycie_jablek_mus = 2,
    zuzycie_miodu_dzem = 5,
    zuzycie_miodu_mus = 5,
    maksimum_jablek = 40,
    maksimum_miodu = 25):
    ### Zdefiniowanie problemu jako maksymalny
    prob = LpProblem("Najlepsza produkcja HURT",LpMaximize)

    # 2 zmienne
    dzem=LpVariable("Dzem",0) # juz tutaj podajemy ograniczenie ze sa nieujemne
    mus=LpVariable("Mus",0)
    # pomocnicze zmienne
    zuzycie_jablek =LpVariable("Jablka",0)
    zuzycie_miodu =LpVariable("Miod",0)

    prob += zysk_z_dzemu * dzem + zysk_z_musu * mus - cena_jablek * zuzycie_jablek - cena_miodu * zuzycie_miodu
    prob += zuzycie_jablek_dzem * dzem + zuzycie_jablek_mus * mus == zuzycie_jablek, "Zuzycie jablek"
    prob += zuzycie_miodu_dzem * dzem + zuzycie_miodu_mus * mus == zuzycie_miodu, "Zuzycie miodu"
    prob += zuzycie_jablek <= maksimum_jablek, "Maksimum jablek"
    prob += zuzycie_miodu <= maksimum_miodu, "Maksimum miodu"

    
    prob.solve()
    # Status wyswietli sie po kompilacji
    print("Status:", LpStatus[prob.status])
    # Optymalny zysk:
    for v in prob.variables():
        print(v.name, "=", v.varValue)
    print("Zysk wyniesie {} zł  ".format( value(prob.objective)))
    
    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob.constraints.items()]
    print("*"*10+"Shadow Prices"+"*"*10)
    print(pd.DataFrame(shadows))
    
    ### Zintegrowanie zmiennych ze sliderami.
interact(hurt_produkcja,cena_jablek = cena_jablek_slider,
    cena_miodu =    cena_miodu_slider,
    zysk_z_dzemu =    zysk_z_dzemu_slider,
    zysk_z_musu =zysk_z_musu_slider,
    zuzycie_jablek_dzem =zuzycie_jablek_dzem_slider,
    zuzycie_jablek_mus =zuzycie_jablek_mus_slider,
    zuzycie_miodu_dzem =zuzycie_miodu_dzem_slider,
    zuzycie_miodu_mus =zuzycie_miodu_mus_slider,
    maksimum_jablek =maksimum_jablek_slider,
    maksimum_miodu =maksimum_miodu_slider)




from ipywidgets import IntSlider

style = {'description_width': 'initial'}
IntSlider(description='A too long description', style=style)




x_widget = FloatSlider(min=0.0, max=10.0, step=0.05)
y_widget = FloatSlider(min=0.5, max=10.0, step=0.05, value=5.0)

def update_x_range(*args):
    x_widget.max = 2.0 * y_widget.value
y_widget.observe(update_x_range, 'value')

def printer(x, y):
    print(x, y)
interact(printer,x=x_widget, y=y_widget)


####Po zmianach dokonywanych suwakiem można zauważyć, że im większa cena np. jabłek, miodu to zysk maleje (rośnie zaś zużycie jabłek i maleje zużycie miodu)
####Zwiekszajac zysk z dzemu/musu zwieksza sie zysk calkowity
####Im wieksze zuzycie jablek_dzem tym bardziej maleje dzem i miod, ale mus sie zwieksza
####Im wieksze zuzycie miodu_dzem tym mniej dzemu, jablek i mniejszy zysk.
#### Im wieksze maksimum_jablek tym wszystko rosnie, oprocz musu
#### Im wieksze makimum_miodu, tym dzem do pewnego momentu wynosi 4.8, a potem maleje do 0, jablka rosna, miod rosnie, miod rosnie, zysk rosnie

## Jak widać poniżej kod jest dużo bardziej rozbudowany ze względu na dodanie sliderów, ale one same ułatwiają pracę.
```
